```python
# function for univariate analysis and plot the bar plot

def survey_analysis(df,column):
    filtered_col=df_survey[column][~df_survey[column].astype(str).str.contains('^no asnwer|^No ans|No Answer|undefined')]
    Values =filtered_col.value_counts()[:5]
    total_count= filtered_col.shape[0]
    print('total number of people who have participated in this question are',total_count,'\n Percentage of people are :',format(Values*100/total_count))
    
    #graph
    fig, ax= plt.subplots()
    sns.barplot(x = Values.index, y = Values.values, ax=ax)
    sns.lineplot(x = Values.index, y = Values.values, color='r', marker='o' ,ax= ax)
    ax.set_xlabel(column, fontsize = 16) # set the font size of the x-axis label
    ax.set_ylabel('counts', fontsize= 16)  # set the font size of the y-axis label
    ax.tick_params(axis='x', labelsize=10, rotation=45) # set the font size of the x-axis tick labels
    ax.tick_params(axis='y', labelsize=14) # set the font size of the y-axis tick labels
    plt.show()
```

```python
def sort_form(df,col):
    filtered_col = df_survey[col][~df_survey[col].astype(str).str.contains('^no asnwer|^No ans|No Answer|undefined')]
    new_df= filtered_col.value_counts().reset_index()[:10]
    new_df.columns = ['Columns','Count']
    new_df['Percentage'] = new_df['Count'].apply(lambda x :round(x*100/filtered_col.shape[0],2))
    print(new_df)
    
    
    #graph
    fig, ax= plt.subplots()
    sns.barplot(x =new_df['Columns'] , y = new_df['Count'], ax=ax)
    sns.lineplot(x =new_df['Columns'] , y = new_df['Count'], color='r', marker='o' ,ax= ax)
    #ax.set_ylabel('Count', fontsize= 16)  # set the font size of the y-axis label
    ax.set_title(col,fontsize = 20)# set title
    ax.tick_params(axis='x', labelsize=10, rotation=90) # set the font size of the x-axis tick labels
    ax.tick_params(axis='y', labelsize=14) # set the font size of the y-axis tick labels
    plt.show()
```

```python
sort_form(df_survey,'Location')
```

```python
cat_col = [values for values in df_survey.columns if df_survey[values].dtype == 'O']
cat_25 = []
for i in cat_col:
    if df_survey[i].value_counts().shape[0] < 25:
        cat_25.append(i)
```

```python
exclude_cols = ['Year', 'Weekday', 'class','Location','id', 'Date', 'Name', 'Month'] # list of columns to exclude

for columns in df_survey.columns:
    if columns not in exclude_cols:
        print(survey_analysis(df_survey,columns))
```

```python
def pie_charts(df, columns):
    filtered_col = df_survey[columns][~df_survey[columns].astype(str).str.contains('^no asnwer|^No Answer|^No answer|undefined')]
    x= filtered_col.value_counts().index[:5]
    y= filtered_col.value_counts().values[:5]
    ## define custom color palette
    
    ##Pie chart
    fig = plt.figure(figsize=(10,10))
    plt.pie(y, autopct= '%1.2f%%', textprops={'fontsize': 20, 'color': 'black'}, shadow=True)
    plt.title(columns, fontsize=20)
    plt.legend(x, fontsize=15, loc='upper left',bbox_to_anchor=(1, 0.5))
    plt.show()
```

```python
exclude_cols = ['Year', 'Weekday', 'class','Location','id', 'Date', 'Name', 'Month'] # list of columns to exclude

for columns in df_survey.columns:
    if columns not in exclude_cols:
        print(pie_charts(df_survey,columns))
```
